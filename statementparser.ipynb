```python
import pandas as pd
import re

# Assuming df is already loaded and 'narration' column exists
df['narration'] = df['narration'].astype(str)

def parse_transaction(transaction_message):
    parsed_data = {
        "Transaction Type": transaction_message,  # Fallback to original narration
        "Name": transaction_message,  # Fallback to original narration
        "UPI ID": transaction_message  # Fallback to original narration
    }

    # Regex pattern to extract UPI ID or similar IDs
    upi_id_pattern = re.compile(r'([a-zA-Z0-9._]+@[a-zA-Z]+)')
    
    # Split the message into components using '-' as a delimiter
    components = transaction_message.split('-')

    # Determine the transaction type (e.g., UPI, IB BILLPAY, IMPS, etc.)
    if "UPI" in components[0]:
        parsed_data["Transaction Type"] = "UPI"
    elif "IB BILLPAY" in components[0]:
        parsed_data["Transaction Type"] = "IB BILLPAY"
    elif "IMPS" in components[0]:
        parsed_data["Transaction Type"] = "IMPS"
    else:
        parsed_data["Transaction Type"] = "Other"
    
    # Extract and assign the name from the narration
    if len(components) > 1:
        parsed_data["Name"] = components[1]
    
    # Extract and assign the UPI ID or similar identifier
    upi_id_match = upi_id_pattern.search(transaction_message)
    if upi_id_match:
        parsed_data["UPI ID"] = upi_id_match.group()

    return parsed_data


# List to hold parsed data
parsed_data_list = []

# Iterate over the 'narration' column and parse the data
for value in df['narration']:
    # Parse the value using the defined function
    parsed_data = parse_transaction(value)
    # Append the parsed data dictionary to the list
    parsed_data_list.append(parsed_data)

# Create a DataFrame from the list of dictionaries
parsed_df = pd.DataFrame(parsed_data_list)

# Now, replace the 'narration' column in the original df with the 'Name' from parsed_df
df['name'] = parsed_df['Name']  # Add 'Name' column to the original DataFrame

# Drop everything except for 'name', 'withdrawal_amt', 'deposit_amt', and 'date'
df = df[['name', 'withdrawal_amt', 'deposit_amt', 'date','closing_balance']]

# Display the modified DataFrame
print(df)

# Optionally, you can save the updated DataFrame to an Excel file
df.to_excel('updated_transactions.xlsx', index=False)
```
